```python
import numpy as np
import pandas as pd
import plotly.express as px
import plotly.io as pio
pio.renderers.default = "notebook_connected"
```

```python
##Data
df = pd.DataFrame(np.random.randn(100,4),columns='A B C D'.split())
```

```python
df_long = df.reset_index().melt(id_vars='index', value_vars=list('ABCD'),
                                var_name='Variable', value_name='Value')
```

```python
fig = px.line(df)
fig.show()
```

```python
fig2 = px.line(df_long,x='index',y='Value',color='Variable',
              title='Line Plot of Columns A-D',
              labels={'index':'Row Index','Value':'Random Value'})
fig2.show()
```

```python
fig_scatter = px.scatter(df,x='A',y='B')
fig_scatter.show()
```

```python
fig_scatter2 = px.scatter(df_long,x="index",y="Value",color="Variable",
                         title="AD over row")
fig_scatter2.show()
```

```python
fig_bar = px.bar(df)
fig_bar.show()
```

```python
fig_bar2 = px.bar(df_long,x="index",y="Value",color="Variable",
                      barmode="group",title="Grouped Bar Plot")
fig_bar2.show()
```

```python
fig_box = px.box(df)
fig_box.show()
```

```python
fig_box2 = px.box(df_long, x="Variable", y="Value", color="Variable",
                 title="Box Plot of Columns A-D")
fig_box2.show()
```

```python
fig_hist = px.histogram(df)
fig_hist.show()
```

```python
fig_hist2 = px.histogram(df_long, x="Value", color="Variable",
                        nbins=30, barmode="overlay",
                        title="Histogram of Values")
fig_hist2.show()
```

```python
fig_area = px.area(df)
fig_area.show()
```

```python
fig_area2 = px.area(df_long, x="index", y="Value", color="Variable",
                   title="Area Plot of Columns A-D")
fig_area2.show()
```

```python
##data
df2 = pd.DataFrame({'x':[1,2,3,4,5],'y':[15,25,35,20,10],
                    'z':[500,400,300,100,200]})
df3 = pd.DataFrame({"category": ["A", "B", "C", "D"],
                    "value": [10, 20, 30, 40],
                    "group": ["X", "X", "Y", "Y"]})
```

```python
fig_pie = px.pie(df3, names="category", values="value")
fig_pie.show()
```

```python
fig_pie2 = px.pie(df2,names='x',values='y',)
fig_pie2.show()
```

```python
fig_sun = px.sunburst(df3, path=["group", "category"], values="value")
fig_sun.show()
```

```python
fig_sun2 = px.sunburst(df2, path=["y", "z"], values="x")
fig_sun2.show()
```

```python
fig_tree = px.treemap(df3, path=["group", "category"], values="value")
fig_tree.show()
```

```python
fig_tree2 = px.treemap(df2, path=["y", "z"], values="x")
fig_tree2.show()
```
