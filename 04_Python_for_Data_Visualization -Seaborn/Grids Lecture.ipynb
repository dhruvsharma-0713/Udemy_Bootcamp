```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
plt.show()
```

```python
tips = sns.load_dataset('tips')
g1 = sns.FacetGrid(data=tips, col='time', row='smoker')
g1.map(sns.histplot, 'total_bill')
plt.show()
```

```python
g1 = sns.FacetGrid(data=tips, col='time', row='smoker')
g1.map(plt.scatter, 'total_bill','tip')
plt.show()
```
